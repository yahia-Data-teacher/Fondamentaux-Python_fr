### Exercice 31. Créer un dictionnaire (dict) `{'posts' : 0, 'pages' : 0}`2. Mettez à jour les "messages" à 1003. Obtenez la valeur de `posts` et ajoutez 10 et stockez-la dans `pages`- ASTUCE : le résultat final devrait être `{'posts' : 100, 'pages' : 110}`

```python
d = {'posts': 0, 'pages': 0}
d
```

```python
d['posts'] = 100
d
```

```python
d['pages'] = d['posts'] + 10
d
```

```python
posts = d['posts']
posts += 10
d['pages'] = posts
d
```
