# Tuples

In Python tuples are very similar to lists, however, unlike lists they are *immutable* meaning they can not be changed. You would use tuples to present things that shouldn't be changed, such as days of the week, or dates on a calendar. 

In this section, we will get a brief overview of the following:

    1.) Constructing Tuples
    2.) Basic Tuple Methods
    3.) Immutability
    4.) When to Use Tuples

You'll have an intuition of how to use tuples based on what you've learned about lists. We can treat them very similarly with the major distinction being that tuples are immutable.

<h3 id="nest">Tuple imbriqué</h3>

Un tuple peut contenir un autre tuple ainsi que d'autres types de données plus complexes. Ce processus est appelé « imbrication ». Considérez le tuple suivant avec plusieurs éléments :

```python
# Create a nest tuple

NestedT =(1, 2, ("pop", "rock") ,(3,4),("disco",(1,2)))
```

Chaque élément du tuple, y compris d'autres tuples, peut être obtenu via un index, comme indiqué sur la figure :

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/TuplesNestOne.png" width="750" align= "centre">

```python
# Print element on each index

print("Element 0 of Tuple: ", NestedT[0])
print("Element 1 of Tuple: ", NestedT[1])
print("Element 2 of Tuple: ", NestedT[2])
print("Element 3 of Tuple: ", NestedT[3])
print("Element 4 of Tuple: ", NestedT[4])
```

Nous pouvons utiliser le deuxième index pour accéder à d'autres tuples, comme illustré dans la figure :

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/TuplesNestTwo.png" width="750" align= "centre">

On peut accéder aux tuples imbriqués :

```python
# Print element on each index, including nest indexes

print("Element 2, 0 of Tuple: ",   NestedT[2][0])
print("Element 2, 1 of Tuple: ",   NestedT[2][1])
print("Element 3, 0 of Tuple: ",   NestedT[3][0])
print("Element 3, 1 of Tuple: ",   NestedT[3][1])
print("Element 4, 0 of Tuple: ",   NestedT[4][0])
print("Element 4, 1 of Tuple: ",   NestedT[4][1])
```

Nous pouvons accéder aux chaînes dans les deuxièmes tuples imbriqués en utilisant un troisième index :

```python
# Print the first element in the second nested tuples

NestedT[2][1][0]
```

```python
# Print the second element in the second nested tuples

NestedT[2][1][1]
```

Nous pouvons utiliser un arbre pour visualiser le processus. Chaque nouvel index correspond à un niveau plus profond dans l'arborescence :

```python
# Print the first element in the second nested tuples

NestedT[4][1][0]
```

```python
# Print the second element in the second nested tuples

NestedT[4][1][1]
```

## Quand utiliser les tuplesVous vous demandez peut-être : "Pourquoi s'embêter à utiliser des tuples alors qu'ils ont moins de méthodes disponibles ?" Pour être honnête, les tuples ne sont pas utilisés aussi souvent que les listes en programmation, mais sont utilisés lorsque l'immuabilité est nécessaire. Si dans votre programme vous faites passer un objet et que vous devez vous assurer qu'il ne soit pas modifié, alors un tuple devient votre solution. Il fournit une source pratique d'intégrité des données.Vous devriez maintenant être capable de créer et d'utiliser des tuples dans votre programmation ainsi que de comprendre leur immuabilité.
