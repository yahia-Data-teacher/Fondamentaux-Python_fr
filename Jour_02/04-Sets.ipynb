# Sets

<h3 id="content">Définir le contenu</h3>

Un ensemble est une collection unique d'objets en Python. Vous pouvez désigner un ensemble par une accolade <b>{}</b>. Python supprimera automatiquement les éléments en double :

```python
# Create a set

set1 = {"pop", "rock", "soul", "hard rock", "rock", "R&B", "rock", "disco"}
set1
```

Le processus de cartographie est illustré dans la figure :

Vous pouvez également créer un ensemble à partir d'une liste comme suit :

```python
# Convert list to set

album_list = [ "Michael Jackson", "Thriller", 1982, "00:42:19", \
              "Pop, Rock, R&B", 46.0, 65, "30-Nov-82", None, 10.0]
album_set = set(album_list)             
album_set
```

Créons maintenant un ensemble de genres :

```python
# Convert list to set

music_genres = set(["pop", "pop", "rock", "folk rock", "hard rock", "soul", \
                    "progressive rock", "soft rock", "R&B", "disco"])
music_genres
```

<h3 id="op">Définir les opérations</h3>

Passons en revue les opérations d'ensemble, car elles peuvent être utilisées pour modifier l'ensemble. Considérons l'ensemble <b>A</b> :

```python
# Sample set

A = set(["Thriller", "Back in Black", "AC/DC"])
A
```

Nous pouvons ajouter un élément à un ensemble en utilisant la méthode <code>add()</code> :

```python
# Add element to set

A.add("NSYNC")
A
```

Si nous ajoutons deux fois le même élément, rien ne se passera car il ne peut y avoir aucun doublon dans un ensemble :

```python
# Try to add duplicate element to the set

A.add("NSYNC")
A
```

Nous pouvons supprimer un élément d'un ensemble en utilisant la méthode <code>remove</code> :

```python
# Remove the element from set

A.remove("NSYNC")
A
```

Nous pouvons vérifier si un élément est dans l'ensemble en utilisant la commande <code>in</code> :

```python
# Verify if the element is in the set

"AC/DC" in A
```

<h3 id="logic">Définit les opérations logiques</h3>

N'oubliez pas qu'avec les ensembles, vous pouvez vérifier la différence entre les ensembles, ainsi que la différence symétrique, l'intersection et l'union :

Considérez les deux ensembles suivants :

```python
# Sample Sets

album_set1 = set(["Thriller", 'AC/DC', 'Back in Black'])
album_set2 = set([ "AC/DC", "Back in Black", "The Dark Side of the Moon"])
```

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/SetsSamples.png" width="650" />

```python
# Print two sets

album_set1, album_set2
```

Comme les deux ensembles contiennent <b>AC/DC</b> et <b>Back in Black</b>, nous représentons ces éléments communs par l'intersection de deux cercles.

<img src = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/SetsLogic.png" width = "650" />

Vous pouvez trouver l'intersection de deux ensembles comme suit en utilisant <code>&</code> :

```python
# Find the intersections

intersection = album_set1 & album_set2
intersection
```

Vous pouvez trouver tous les éléments qui ne sont contenus que dans <code>album_set1</code> en utilisant la méthode <code>difference</code> :

```python
# Find the difference in set1 but not set2

album_set1.difference(album_set2)  
```

Vous n'avez qu'à prendre en compte les éléments de <code>album_set1</code> ; tous les éléments de <code>album_set2</code>, y compris l'intersection, ne sont pas inclus.

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/SetsLeft.png" width="650" />

Les éléments dans <code>album_set2</code> mais pas dans <code>album_set1</code> sont donnés par :

```python
album_set2.difference(album_set1)  
```

<img src = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/SetsRight.png" width="650" />

Vous pouvez également trouver l'intersection de <code>album_list1</code> et <code>album_list2</code>, en utilisant la méthode <code>intersection</code> :

```python
# Use intersection method to find the intersection of album_list1 and album_list2

album_set1.intersection(album_set2)   
```

Cela correspond à l'intersection des deux cercles :

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/SetsIntersect.png" width="650" />

L'union correspond à tous les éléments des deux ensembles, ce qui est représenté en colorant les deux cercles :

<img src = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/SetsUnion.png" width="650" />

L'union est donnée par :

```python
# Find the union of two sets

album_set1.union(album_set2)
```

Et vous pouvez vérifier si un ensemble est un sur-ensemble ou un sous-ensemble d'un autre ensemble, respectivement, comme ceci :

```python
# Check if superset

set(album_set1).issuperset(album_set2)   
```

```python
# Check if subset

set(album_set2).issubset(album_set1)     
```

Voici un exemple où <code>issubset()</code> et <code>issuperset()</code> renvoient true :

```python
# Check if subset

set({"Back in Black", "AC/DC"}).issubset(album_set1) 
```

```python
# Check if superset

album_set1.issuperset({"Back in Black", "AC/DC"})   
```
