# Project: LEET

Leet (or "1337"), also known as eleet or leetspeak, is a system of modified spellings used primarily on the Internet. ([wiki](https://en.wikipedia.org/wiki/Leet))

### Description du projet
- Nous prenons une chaîne d'entrée de l'utilisateur et la convertissons en un leet simplifié.


### Lettre simplifiée
- Nous ne convertirons que certaines lettres.

#### Leet table

| lettre | laisser |
| --- | --- |
| Un | 4 |
| Et | 3 |
| est | 1 |
| L | 1 |
| Le | 0 |
| S | 5 |
| T | 7 |


### Étape 1 : Entrée de l'utilisateur

Demander à l'utilisateur une phrase / un mot à convertir en leet


```python
sentence = input('What should I convert? ')
```

```python
sentence
```

### Étape 2 : Convertir en majuscules

Convertissez l'entrée de l'utilisateur en majuscules.


```python
sentence_upper = sentence.upper()
```

```python
sentence_upper
```

### Étape 3 : Pour chaque lettre du tableau ci-dessus, remplacez-la par leet.


```python
sentence_upper.replace('A', '4').replace('E', '3').replace('I', '1').replace('L', '1').replace('O', '0').replace('S', '5').replace('T', '7')
```
