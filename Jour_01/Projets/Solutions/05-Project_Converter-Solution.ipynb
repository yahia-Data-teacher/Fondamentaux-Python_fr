# Project Converter

Create a converter from fahrenheit to celcius using the formula

$$ °C = (°F − 32) \times 5/9 $$

### Étape 1 : saisie

Demandez à l'utilisateur d'entrer une température en Fahrenheit.


```python
fahrenheit_str = input('Input temperature in Fahrenheit: ')
```

### Étape 2 : Convertir

Convertir l'entrée en `float`


```python
fahrenheit = float(fahrenheit_str)
```

```python
type(fahrenheit), fahrenheit
```

### Étape 3 : Calculer

Utilisez la formule pour calculer la température en degrés Celsius


```python
celcius = (fahrenheit - 32)*5/9
```

### Étape 4 : Imprimez

Imprimer le résultat à l'utilisateur


```python
print('Fahrenheit', fahrenheit)
print('Celcius', celcius)
```
