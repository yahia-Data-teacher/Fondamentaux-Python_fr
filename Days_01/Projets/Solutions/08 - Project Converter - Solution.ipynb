# Project Converter

Create a converter from fahrenheit to celcius using the formula

$$ °C = (°F − 32) \times 5/9 $$

### Step 1: input

Ask the user to input a temperature in Fahrenheit.

```python
fahrenheit_str = input('Input temperature in Fahrenheit: ')
```

### Step 2: Convert

Convert the input to a `float`

```python
fahrenheit = float(fahrenheit_str)
```

```python
type(fahrenheit), fahrenheit
```

### Step 3: Calculate

Use the formula to calculate the temperature in celcius

```python
celcius = (fahrenheit - 32)*5/9
```

### Step 4: Print

Print the result to the user

```python
print('Fahrenheit', fahrenheit)
print('Celcius', celcius)
```
