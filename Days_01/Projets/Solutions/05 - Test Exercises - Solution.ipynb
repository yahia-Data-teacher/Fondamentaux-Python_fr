### Exercise 3
You need to help a farmer. He asks you to tell him how many legs can be counted among all his animals.

He has 3 species.
- Chickens: 2 legs
- Cows: 4 legs
- Sheeps: 4 legs

1. Ask the user to input the number of chickens
2. Ask the user to input the number of cows
3. Ask the user to input the number of sheeps
4. Calculate how many legs there are in total.

#### DOUBLE CLICK FOR HINT

<!--

-->

```python
chickens = int(input('How many Chickens? '))
cows = int(input('How many Cows? '))
sheeps = int(input('How many Sheeps? '))
```

```python
print('Number of legs', chickens*2 + cows*4 + sheeps*4)
```

```python
chickens_str = input('How many Chickens? ')
chickens = int(chickens_str)

cows_str = input('How many Cows? ')
cows = int(cows_str)

sheeps_str = input('How many Sheeps? ')
sheeps = int(sheeps_str)

number_of_legs = chickens*2 + cows*4 + sheeps*4
print('Number of legs', number_of_legs)
```
